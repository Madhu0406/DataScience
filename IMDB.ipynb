```python
#Establishing the connection with the databse
cur.execute("SELECT * FROM movies")
#Creating the cursor Object
movies = cur.fetchall()
#Creating a Dataframe
movies = pd.DataFrame(movies,columns=['id', 'Original_title' , 'Budget', 'Popularity' , 'Relase_date' , 'Revenue' , 'Title' , 'Avg_Rating' ,'Rating_count' , 'Overview' , ' Tagline' , 'UID' , 'Director_ID'])
```
